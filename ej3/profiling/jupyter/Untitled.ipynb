```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('../results/density.csv')
```

```python
plt.clf()
axis1 = plt.subplot(2,1,1)
axis2 = plt.subplot(2,1,2)
df.plot(x='density', y='time', kind='box', ax=axis1)
df.plot(x='density', y='time', kind='density', ax=axis2)
plt.show()
```

```python
df = pd.read_csv('../results/n_growth.csv')
```

```python
plt.clf()
ax = df.plot(x='n', logy=True)
ax.legend(['Crecimiento de N'])
ax.set_ylabel('segundos (log)')
ax.set_xlabel('n')
plt.show()
```
