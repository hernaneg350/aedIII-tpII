```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('../results/high_vs_low_density.csv')
```

```python
plt.clf()
ax = df.plot(x='n', logy=True)
ax.legend(['Alta densidad', 'Baja densidad'])
ax.set_ylabel('segundos (log)')
ax.set_xlabel('tamano de instancia')
plt.show()
```
